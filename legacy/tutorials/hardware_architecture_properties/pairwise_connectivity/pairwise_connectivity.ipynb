## Pairwise connectivity

```python
# Relative path to _helpers
import sys
import pathlib
import os
sys.path.insert(0, str(pathlib.Path(os.path.abspath('')).parent.parent.parent.resolve()))


# Disregard any warning on the following imports
from _helpers.circuit_submitter import CircuitSubmitter
from qiskit import QuantumCircuit


device_name = "noisy_sim"

submitter = CircuitSubmitter("number of usable qubits", device_name)
```

```python
# Put here the connectivities provided by the hardware manufacturer
# Note the directionality, where the first number in each tuple acts as control and the second as target
reported_connectivities = [(0, 1), (1, 0), (0, 2), (2, 0), (1, 2)]
```

```python
for control, target in reported_connectivities:
    qc = QuantumCircuit(max(control, target) + 1, max(control, target) + 1)
    qc.cx(control, target)
    qc.measure_all()

    try:
        submitter.submit_circuits(shots=1, verbatim=True, skip_asking=True, qasm_strs=[qc.qasm()])
        submitter.retrieve_counts()
        print(f"There were no errors during the execution. The directional connectivity {control} -> {target} is valid.\n")
    except Exception as e:
        print(e)
        print(f"Exception catched! The directional connectivity {control} -> {target} is not valid.\n")
```
