## Number of usable qubits

```python
# Relative path to _helpers
import sys
import pathlib
import os
sys.path.insert(0, str(pathlib.Path(os.path.abspath('')).parent.parent.parent.resolve()))


# Disregard any warning on the following imports
from _helpers.circuit_submitter import CircuitSubmitter
from qiskit import QuantumCircuit

device_name = "noisy_sim"

submitter = CircuitSubmitter("number of usable qubits", device_name)
```

```python
# Put here the number of qubits provided by the hardware manufacturer
reported_n_qubits = 10
```

```python
usable_qubits = reported_n_qubits

for num_qubits in range(1, reported_n_qubits + 1):
    try:
        print(f"Testing {num_qubits} qubits:")
        qc = QuantumCircuit(num_qubits, num_qubits)
        for i in range(num_qubits):
            qc.x(i)
        qc.measure_all()
        submitter.submit_circuits(shots=1, verbatim=False, skip_asking=True, qasm_strs=[qc.qasm()])
        submitter.retrieve_counts()
        usable_qubits = num_qubits
        print(f"There were no errors during the execution. The device has at least {usable_qubits} usable qubits.\n")
    except Exception as e:
        print(e)
        print(f"Exception catched! The device does not have {usable_qubits} usable qubits.\n")
        break

print(f"The number of usable qubits is {usable_qubits}.")
```
